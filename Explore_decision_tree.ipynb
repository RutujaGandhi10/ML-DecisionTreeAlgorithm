
### Exploring Decision Tree Classifier And Visualizing the tree in different ways, 
### to predict the right class for new data with the simple but famous dataset - "Iris"     

   

```python
# Let's load the necessary libraries and packages
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

```python
#Let's load the iris data
iris = datasets.load_iris()
```

```python
#Let's prepare the data
X = iris.data
y = iris.target
```

```python
# Fit the classifier 
DT = DecisionTreeClassifier()
model = DT.fit(X, y)
```

Here, our model is ready. Let's visualize the decision tree.

#### Let's visualize the decision tree in 4 different ways.

#### 1. By printing text representation of the tree with sklearn.tree.export_text method

#### 2. Plotting the tree plot with sklearn.tree.plot_tree method(matplotlib)

#### 3. Plotting the graph with sklearn.tree.export_graphviz method(graphviz)

#### 4. Plotting the tree graph with dtreeviz package(dtreeviz)



1. By printing text representation of the tree with sklearn.tree.export_text method

```python
#text representation 
DT_text_representation = tree.export_text(DT)
print(DT_text_representation)
```

```python
# Save the text representation in file
with open("decistion_tree.log", "w") as fout:
    fout.write(DT_text_representation)
```

Now, Let's go for the second visualization technique

2. Plotting the tree plot with sklearn.tree.plot_tree method(matplotlib)

```python
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(DT, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
fig.savefig("decistion_tree.png")
```

Let's go for the third visualization technique

3. Plotting the graph with sklearn.tree.export_graphviz method(graphviz)

```python
import os
os.environ["PATH"]+= os.pathsep+(r'C:\Program Files (x86)\Graphviz2.38\bin')
import graphviz
# DOT data
dot_data = tree.export_graphviz(DT, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True,rounded=True,special_characters=True)
# Draw graph
Tree_graph = graphviz.Source(dot_data,format="png") 
Tree_graph
```

```python
#Let's save this in png format

Tree_graph.render("decision_tree_graphivz")
```

Let's go for the fourth visualization technique

4. Plotting the tree graph with dtreeviz package(dtreeviz)

```python
from dtreeviz.trees import dtreeviz # remember to load the package
viz = dtreeviz(DT, X, y, target_name="target", feature_names=iris.feature_names, class_names=list(iris.target_names))
viz
```

```python
# Let's predict on some custom input values.
```

```python
# Let's take an array of categories for user friendly prediction at the place of 0,1,2
category = ['Iris-Setosa','Iris-versicolor','Iris-virginica']
```

```python
new_DT=list([[5.7 ,4.4, 1.5, 0.4]])
new_DT_prediction=model.predict(new_DT)
new_DT_prediction[0]
print(category[int(new_DT_prediction[0])])
```

```python
neww_DT=list([[6.1, 3, 4.6, 1.4]])
neww_DT_prediction=model.predict(neww_DT)
neww_DT_prediction[0]
print(category[int(neww_DT_prediction[0])])
```
